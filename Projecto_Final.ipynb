## 10.2 Tratamiento de valores nulos

Funciones utilizadas:
- dropna(subset=[...]): elimina filas con nulos en columnas críticas.
- fillna(valor): imputa nulos con un valor (p.ej., mediana para numéricos, moda para categóricos).
- mode()/median(): moda/mediana de la columna.
- len(df): Guarda cuántas filas había antes de limpiar
- reset_index(drop=True): reorganiza los índices después de eliminar filas.
Imprime cuántas filas se eliminaron.

```python
# Eliminar nulos criticos cantidad.
if "cantidad" in df.columns:
  df = df.dropna(subset=["cantidad"])

# Imputar columnas numéricas con la mediana
for col in ["cantidad"]:
  if col in df.columns and df[col].isna().sum() > 0:
    df[col] = df[col].fillna(df[col].median())

# Imputar columnas categóricas con la moda
for col in ["fecha", "producto","categoria","almacen","movimiento"]:
  if col in df.columns and df[col].isna().sum() > 0:
    df[col] = df[col].fillna(df[col].mode(dropna=True)[0])

# Verificación final
print(f"Nulos restantes totales: {df.isnull().sum().sum()}")
```

##10.3 Normalización y tipos

Funciones utilizadas:
- pd.to_numeric(..., errors="coerce"): convierte a numérico, y si algo no se puede convertir lo pone NaN (útil para limpiar).
- astype(str).str.strip().str.lower(): normaliza categóricas (texto uniforme sin espacios y en minúsculas).
- dtypes: verifica tipos actuales.
- pd.to_datetime(..., errors="coerce"): convierte a fecha y *hora*, y si algo no se puede convertir lo pone NaN (útil para limpiar).

```python
#para manipular caracteres Unicode (quitar acentos).
import unicodedata
#Función para quitar acentos
def strip_accents(text):
    if isinstance(text, str):
        return ''.join(
            char for char in unicodedata.normalize('NFD', text)
            if unicodedata.category(char) != 'Mn'
        )
    return text

if "cantidad" in df.columns:
  #Convertir columna cantidad a numerico
  df["cantidad"] = pd.to_numeric(df["cantidad"], errors="coerce")
else:
 # Genérico: toma numéricas detectadas automáticamente
  cols_num = df.select_dtypes(include=["number"]).columns.tolist()
  print("No se detectaron columnas numéricas esperadas.")
cols_num = [c for c in ["cantidad"] if c in df.columns]
#Convertir columna fecha a formato date
if "fecha" in df.columns:
  df["fecha"] = pd.to_datetime(df["fecha"], errors="coerce")

#Aplica las siguientes varias limpiezas a las columnas
for col in ["producto","categoria","almacen","movimiento"]:
  if col in df.columns:
    df[col] = df[col].astype(str).str.strip().str.lower().apply(strip_accents)

print("---- Primeras filas ----")
print("="*105)
print(df.head().to_markdown(index=False))
print("="*105)

print("\n---- Tipos de datos ----")
print("="*105)
print(df.dtypes)
print("="*105 + "\n")
```

## 10.4 Tratamiento de duplicados

Funciones utilizadas:
- duplicated(): marca filas repetidas.
- drop_duplicates(): elimina duplicados y deja solo una copia.
- reset_index(drop=True): Resetea los índices.

```python
# Eliminar duplicados exactos
dup_count = df.duplicated().sum()
if dup_count > 0:
    df = df.drop_duplicates().reset_index(drop=True)
    print(f"Se eliminarán {dup_count} filas duplicadas exactas.")
else:
    print("No hay duplicados exactos.\n")

print(f"Total de filas después de la limpieza: {len(df)}\n")
```

## 10.5 Validación de rangos y correcciones

Funciones utilizadas:
- clip(lower=..., upper=...): recorta valores fuera de rango.
- Filtros condicionales: verificar reglas de negocio (p.ej., cantidad ≥ 0, id_inventario >=1).
- pd.Timestamp.today(): Obtiene la fecha y hora actual del sistema, en formato timestamp (con fecha + hora).
- pd.to_datetime(): Convierte un dato (string o número) en fecha (tipo datetime).
- isin(): Revisa si el valor de una columna pertenece a una lista definida.

```python
import pandas as pd

def obtener_estacion(fecha):
    # Definir las estaciones para el hemisferio sur como ejemplo
    # Ajusta estas fechas si el contexto es Hemisferio Norte u otra convención
    if (fecha.month == 12 and fecha.day >= 21) or \
       (fecha.month == 1) or \
       (fecha.month == 2) or \
       (fecha.month == 3 and fecha.day < 21):
        return 'verano'
    elif (fecha.month == 3 and fecha.day >= 21) or \
         (fecha.month == 4) or \
         (fecha.month == 5) or \
         (fecha.month == 6 and fecha.day < 21):
        return 'otono'
    elif (fecha.month == 6 and fecha.day >= 21) or \
         (fecha.month == 7) or \
         (fecha.month == 8) or \
         (fecha.month == 9 and fecha.day < 21):
        return 'invierno'
    else: # (fecha.month == 9 and fecha.day >= 21) or (fecha.month == 10) or (fecha.month == 11) or (fecha.month == 12 and fecha.day < 21)
        return 'primavera'

# Asegurarse de que la columna 'fecha' sea de tipo datetime
# Nota: La columna 'fecha_reconstructed' se eliminó en el paso anterior, por lo que usaremos la original 'fecha'
# Si 'fecha' no existiera o estuviera en formato incorrecto, necesitaríamos el código del paso 10.3 para convertirla.
if 'fecha' in df.columns and not pd.api.types.is_datetime64_any_dtype(df['fecha']):
    df['fecha'] = pd.to_datetime(df['fecha'], errors='coerce')

# Aplicar la función para crear la nueva columna 'estacion'
df['estacion'] = df['fecha'].apply(obtener_estacion)

print("DataFrame con la nueva columna 'estacion':")

display(df.head())
```

```python
import pandas as pd

# Diccionario para guardar resultados
resultados = {}

# 1. Relación entre temporada del año y movimiento de productos
if "estacion" in df.columns and "categoria" in df.columns:
    resultados["movimiento_por_estacion_y_categoria"] = (
        df.groupby(["estacion", "categoria"])["cantidad"]
          .agg(["sum", "mean", "count"])

    )

# 2. Variación de cantidad de productos por categoría
if "categoria" in df.columns:
    resultados["cantidad_por_categoria"] = pd.crosstab(
        index=df["categoria"],
        columns="Total de registros"
    )

# 3. Frecuencia de entradas y salidas de productos por almacén
if "almacen" in df.columns and "movimiento" in df.columns:
    resultados["entradas_vs_salidas_por_almacen"] = pd.crosstab(
        df["almacen"], df["movimiento"]
    )

# 4. Tendencia de escasez o exceso de stock a lo largo del tiempo (por mes)
if "fecha" in df.columns:
    # Tendencia de stock por mes (sin crear columna nueva en df)
    resultados["tendencia_stock_por_mes"] = (
        df.groupby(df['fecha'].dt.to_period('M'))['cantidad'].sum().reset_index()
        .rename(columns={'fecha': 'mes_anio', 'cantidad': 'total_cantidad'})
    )

    # Top 10 días con mayor cantidad movida
    total_por_dia = df.groupby("fecha")["cantidad"].sum().reset_index()
    top10_dias = total_por_dia.sort_values(by="cantidad", ascending=False).head(10)
    resultados["top10_dias_mayor_movimiento"] = top10_dias

# Mostrar todos los resultados generados
for nombre, tabla in resultados.items():
    print(f"\n=== RESULTADO: {nombre} ===")
    display(tabla)
```

```python
import pandas as pd
import sys
from IPython.display import display

# 1. Separar Campos en Producto
if "producto" in df.columns:
    # Extraer la parte después del guion como id_inventario
    df['id_inventario'] = df['producto'].astype(str).str.split('-').str[1]

    # Convertir a numérico, coaccionando errores a NaN
    df['id_inventario'] = pd.to_numeric(df['id_inventario'], errors='coerce')

    # Extraer la parte antes del guion como el nombre del producto
    df['producto_nombre_temp'] = df['producto'].astype(str).str.split('-').str[0].str.strip()

    # Eliminar la columna original 'producto'
    df = df.drop(columns=['producto'])

    # Renombrar la columna temporal al nombre final 'producto'
    df = df.rename(columns={'producto_nombre_temp': 'producto'})
else:
    print("\n[Paso 0] Advertencia: La columna 'producto' no se encontró para la extracción de ID.")
df["id_inventario"] = pd.to_numeric(df["id_inventario"], errors="coerce")

# 2. Bins por RANGOS MANUALES de "cantidad"
if "cantidad" in df.columns:
    df["cantidad_rango"] = pd.cut(
        df["cantidad"],
        bins=[0, 7, 14, 20],
        labels=["Bajo (1-7)", "Medio (8-14)", "Alto (15-20)"],
        include_lowest=True, # Incluir el límite inferior (0 en este caso)
        right=True          # El intervalo es (bin_i, bin_{i+1}]
    )
else:
    print("[Paso 2] Advertencia: La columna 'cantidad' no se encontró para el binning manual.")

# 3. Bins por CUANTILES (Q1, Q2, Q3, Q4) de "cantidad"
if "cantidad" in df.columns:
    df["cantidad_quantil"] = pd.qcut(
        df["cantidad"],
        q=4, # Dividir en 4 cuantiles (cuartiles)
        labels=["Q1 (Muy Bajo)", "Q2 (Bajo)", "Q3 (Alto)", "Q4 (Muy Alto)"],
        # Manejar valores duplicados si hay muchos ceros o el mismo valor en los puntos de corte
        duplicates='drop'
    )
else:
    print("[Paso 3] Advertencia: La columna 'cantidad' no se encontró para el binning por cuantiles.")

# 4. Tabla resumen para visualizar resultados (INCLUYE CAMPOS ADICIONALES)
print("\nVariables derivadas creadas correctamente. Mostrando las primeras 5 filas (incluyendo campos originales):")

# Definir las columnas a mostrar, asegurando que existan en el DataFrame final
display_cols = [
    col for col in [
        "id_inventario", "producto", "categoria", "almacen", "movimiento",
        "fecha", "cantidad",
        "cantidad_rango", "cantidad_quantil","estacion"
    ]
    if col in df.columns
]

# Muestra el encabezado del DataFrame con las columnas transformadas
display(df[display_cols].head())

print("\nTipo de datos de las nuevas columnas:")
# Muestra los tipos de datos de las columnas mostradas
print(df[display_cols].dtypes)
```
